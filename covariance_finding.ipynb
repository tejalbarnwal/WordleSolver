# Relation between two letters in the same word

## Importing Libraries

```python
import numpy as np
import pandas as pd
import string
import itertools
```

## Defining Functions

```python
#defining functions

#splitting word into letters
def split(word):
    return [char for char in word]

#checking if word has two letters together
def has_both_letters(pair, word):
    if (pair[0]!=pair[1]): # case of different letters
      if pair[0] in word and pair[1] in word:
          return True
      else:
          return False
    else: # case of same letters
      count = 0
      for i in word:
        if i == pair[0]:
            count += 1
      if(count>=2):
        return True
      else:
        return False
```

```python
solution_list = list(solutions["word"])
```

```python
alphabet_list = list(string.ascii_uppercase)
# alphabet_list
```

```python
pair_list = list(itertools.product(alphabet_list,alphabet_list))
# pair_list
```

## Finding the covaraince relation

```python
df_covariance = pd.DataFrame(columns = alphabet_list, index = alphabet_list)
```

```python
for pair in pair_list:
  count = 0
  for word in solution_list:
    count = has_both_letters(pair, word) + count
  df_covariance.loc[pair[0]].at[pair[1]] = count
```
